```python
import pandas as pd
import time
```

```python
bus_stops_df = pd.read_csv("bus_stops.csv")
bus_stops_df.head()
```

```python
name=list(bus_stops_df['RoadName'])
desc=list(bus_stops_df['Description'])
code=list(bus_stops_df['BusStopCode'])

l=[[code[i],name[i]+','+desc[i]] for i in range(len(code))]
print(l[:10])
```

```python
size=4099
prime=17

def h(x): 
    return x%size
def d(x):  #d=hashIncr(k)
    return x%prime+1
def reh1(j,d,i):
    return (j+i*d)%size
def reh2(j,d,i):
    return (j-i*d)%size
def reh3(j,d,i):
    return (j-i**2*d)%size
'''
def loadf(ht):
    a,b=0,0
    for i in ht:
        if ht[i]!=-1:
            a+=1
        b+=1
    return a/b
'''

def search(ht,key,reh):
        
    i=1 #i-1 is the key comparison(collision)
    k=h(key)  #h(k)
    while True:
        if ht[k][0]==-1:   #value -1 means empty
            return False,k,i-1,ht[k][1]   #return position k for insertion, ht[k][1] is the info
        if ht[k][0]==key:  #found
            return True,k,i-1,ht[k][1]
        
        #collision
        k=reh(h(key),d(key),i) 
        i+=1
        
def insert(ht,key,info,reh):

    found,pos,collision,info2=search(ht,key,reh)
    
    ht[pos][0],ht[pos][1]=key,info
    #return found,pos,collision,info
```

```python
def storeAndSearch(loadfactor,reh):

    ht={i:[-1,''] for i in range(size)} #create the empty hashtable
    
    #store size*loadfactor number of key values into the hashtable
    for i in range(int(size*loadfactor)):
        insert(ht,l[i][0],l[i][1],reh)

    totaltime=0
    for k in range(10): #repeat 10 times to calculate the average CPU time
        unsucc,succ=[0,0],[0,0] #succ[0] means the total number of successful search, succ[1] means the total collisions of successful search
        start=time.time() #start time
        for i in l:   #search for all the data in the dataset
            found,pos,collision,info=search(ht,i[0],reh)
            if not found:   #i is not inside the table
                unsucc[0]+=1
                unsucc[1]+=collision
            else:
                succ[0]+=1
                succ[1]+=collision
                #if collision!=0:
                    #print(i,collision)
        totaltime+=time.time()-start
        
    print('successful search:',succ[0],'  ',succ[1],'comparisons')
    print('unsuccessful search:',unsucc[0],'  ',unsucc[1],'comparisons')
    
    print('ave CPU time:',round(totaltime/10,4))
    
    print('ave key comparisons (successful): ',round(succ[1]/succ[0],4))
    print('ave key comparisons (unsuccessful): ',round(unsucc[1]/unsucc[0],4))
    print('ave key comparisons in total: ',round((succ[1]+unsucc[1])/(succ[0]+unsucc[0]),4))
    
    return ht
```

## Load Factor=0.25

```python
ht=storeAndSearch(0.25,reh1)
```
