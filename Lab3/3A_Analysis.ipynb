# Random

```python
random = pd.DataFrame()
random = random.append(df_10power3.iloc[0])
random = random.append(df_10power4.iloc[0])
random = random.append(df_10power5.iloc[0])
random = random.append(df_250000.iloc[0])
random = random.append(df_500000.iloc[0])
random = random.append(df_750000.iloc[0])
# random = random.append(df_10power6.iloc[0])
random.rename(columns={"Unnamed: 0" : "input_size"}, inplace=True)
random = random.reset_index(drop=True)
idx=0
for i in range(3,6):
    random.at[idx, "input_size"] = pow(10, i)
    idx+=1
random.at[3, "input_size"] = 250000
random.at[4, "input_size"] = 500000
random.at[5, "input_size"] = 750000
# random.at[7, "input_size"] = 1000000
random
```

```python
f, axes = plt.subplots(1,2,figsize=(18,9))
axes[0].scatter(x="input_size", y="Key Comp - ISort", marker="x", data=random)
axes[0].scatter(x="input_size", y="Key Comp - MSort", marker="x", data=random)
axes[0].set_xlabel("input_size")
axes[0].set_ylabel("Key Comparisons")
axes[0].legend()

axes[1].scatter(x="input_size", y="ISort Time", marker="x", data=random)
axes[1].scatter(x="input_size", y="MSort Time", marker="x", data=random)
axes[1].set_xlabel("input_size")
axes[1].set_ylabel("Time taken");
axes[1].legend()
plt.show()
```

```python
print("Random dataset:")
f, axes = plt.subplots(2,2,figsize=(18,9))
axes[0][0].scatter(x='input_size', y='ISort Time', data=random)
axes[0][1].scatter(x='input_size', y='MSort Time', data=random)
axes[0][0].legend()
axes[0][1].legend()

axes[1][0].scatter(x='input_size', y='Key Comp - ISort', data=random)
axes[1][1].scatter(x='input_size', y='Key Comp - MSort', data=random)
axes[1][0].legend()
axes[1][1].legend()
plt.show()
```
